```python
import pandas as pd

# constants
tokens = ["TCS.NS", "^NSEI", "^NSEBANK", "INFY.NS", "HINDUNILVR.BO",'RIGDL.XC']
token = tokens[-1]
duration = 60
test_len = 300
# get_data(token, duration)
df = pd.read_csv('DABUR_with_indicators_.csv')
```

```python
df=df[['Datetime','Open','High','Low','Close']]
df['Datetime']=pd.to_datetime(df['Datetime'])
df.head(10)
```

```python
#add day of  features
df['day_week']=df['Datetime'].dt.dayofweek
df['day_month']=df['Datetime'].dt.day
df['month']=df['Datetime'].dt.month

df['time']=df['Datetime'].map(lambda a:(pd.Timedelta(a-pd.to_datetime('2023-1-14 9:15:00+05:30'))/pd.Timedelta('5m'))%(24*12))
```

```python
test_len=1200
```

```python
# dividing test and train
train = df[:-test_len]
test = df[-test_len:]
```

```python
# making target 
train['Open_lag']=train.copy()['Open'].shift(-3)
test['Open_lag']=test.copy()['Open'].shift(-3)
```

```python
test
```

```python
train
```

```python
train['Tar']=train['Open_lag']-train['Open']
train['Tar']=train['Tar'].map(lambda a : int(a>0))
test['Tar']=test['Open_lag']-test['Open']
test['Tar']=test['Tar'].map(lambda a : int(a>0))
```

```python
test
```

```python
# functions to make features

def make_lag(label,lags):
    for i in lags:
        print(f"'{label}_lag_{i}',",end='')
        test[f'{label}_lag_{i}']=test[label].shift(i)
        train[f'{label}_lag_{i}']=train[label].shift(i)

def moving_avg(label,lags):
    for i in lags: 
        print(f"'{label}_moving_avg_{i}',",end='')
        train[f'{label}_moving_avg_{i}']=train[label].rolling(window=i, center=False).mean()
        test[f'{label}_moving_avg_{i}']=test[label].rolling(window=i, center=False).mean()
```

```python
make_lag('Open',[6,3])
make_lag('Close',[6,3])
# make_lag('High',[6,12,3])
# make_lag('Low',[6,12,3])

moving_avg('Open',[6,3])
# moving_avg('Close',[6,12,3])
# moving_avg('High',[6,12,3])
# moving_avg('Low',[6,12,3])
```

```python
# Final Data for the model
features = [
    "Open",
    "High",
    "Low",
    "Close",
    "day_week",
    "day_month",
    "month",
    "time",
    "Open_lag_6",
    # "Open_lag_12",
    "Open_lag_3",
    "Close_lag_6",
    "Close_lag_3",
    # "High_lag_6",
    # "High_lag_12",
    # "High_lag_3",
    # "Low_lag_6",
    # "Low_lag_12",
    # "Low_lag_3",
    "Open_moving_avg_6",
    # "Open_moving_avg_12",
    "Open_moving_avg_3",
    # "Close_moving_avg_6",
    # "Close_moving_avg_12",
    # "Close_moving_avg_3",
    # "High_moving_avg_6",
    # "High_moving_avg_12",
    # "High_moving_avg_3",
    # "Low_moving_avg_6",
    # "Low_moving_avg_12",
    # "Low_moving_avg_3",
]
train = train.dropna(axis=0)
test = test.dropna(axis=0)
X_train = train[features]
y_train = train["Tar"]
X_test = test[features]
y_test = test["Tar"]
```

```python
# Making model 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
correct = 0
super_correct = 0
predictions = list(model.predict(X_test))
wrong = 0
confidence =0.45
for i in range(len(predictions)):
    if (predictions[i] > 0.5 and list(y_test)[i] == 1) or (
        predictions[i] < 0.5 and list(y_test)[i] == 0
    ):
        correct += 1
    if (predictions[i] > 0.5 + confidence and list(y_test)[i] == 1) or (
        predictions[i] < 0.5 - confidence and list(y_test)[i] == 0
    ):
        super_correct += 1
    if (predictions[i] > 0.5 + confidence and list(y_test)[i] == 0) or (
        predictions[i] < 0.5 - confidence and list(y_test)[i] == 1
    ):
        wrong += 1
print(f'Total { correct} out of {len(predictions)} are correct')
print(f'Accuracy is {correct/len(predictions)*100} % ')
print(f'Total { super_correct} out of {super_correct+wrong} are super correct')
print(f'Super accuracy is {super_correct*100/(super_correct+wrong)}')



```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(list(y_test),predictions)
```
